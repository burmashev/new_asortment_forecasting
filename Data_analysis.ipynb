### Firstly, we need to download and convert train and test dataset. ###

```python
import pandas as pd
train = pd.read_csv('/Users/alexander/АДВБ/hse-dab-naf-23/train.csv', sep=',')
train
```

```python
test = pd.read_csv('/Users/alexander/АДВБ/hse-dab-naf-23/test.csv')
test
```

```python
test = train[train['id'].isin(test['id'].tolist())].copy()
test
```

```python
train = train[train['demand'].notna()].copy()
train
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
import numpy as np
flags = ['PROMO1_FLAG', 'PROMO2_FLAG', 'AUTORIZATION_FLAG']
fill_mean = ['PRICE_REGULAR', 'PRICE_AFTER_DISC', 'NUM_CONSULTANT']

for column in flags:
    train[column].fillna(0, inplace=True)
    test[column].fillna(0, inplace=True)
for column in fill_mean:
    train[column].fillna(np.mean(train[column]), inplace=True)
    test[column].fillna(np.mean(test[column]), inplace=True)
```

```python
train.dtypes
```

```python
import datetime as dt
train['period_dt'] = pd.to_datetime(train['period_dt'])
test['period_dt'] = pd.to_datetime(test['period_dt'])
train['period_dt'] = train['period_dt'].apply(lambda x: x.date())
test['period_dt'] = test['period_dt'].apply(lambda x: x.date())
integers = ['location_id', 'product_id', 'PROMO1_FLAG', 'PROMO2_FLAG', 'NUM_CONSULTANT', 'AUTORIZATION_FLAG']
for column in integers:
    train[column] = train[column].astype(int)
    test[column] = test[column].astype(int)
```

```python
train.sort_values(by=['period_dt'], inplace=True)
train
```
